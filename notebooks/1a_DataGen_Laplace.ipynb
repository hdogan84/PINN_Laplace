```python
# Analytical solution for Grad(p).n = 0
# p(x,y) = x^2 - y^2
```

```python

len_x = 16
len_y = 16
```

Sample Laplace solution

```python
def target_func(x, y): 
    pres = x*x - y*y
    return pres
```

TRAIN Data

```python
x_coor = np.linspace(0,1,len_x)
y_coor = np.linspace(0,1,len_y)
```

```python
train_dataset = []
for x in x_coor:
    for y in y_coor:
        train_dataset.append((x,y,target_func(x,y)))
```

```python
df_train = pd.DataFrame(train_dataset, columns=['x','y','p(x,y)'])
df_train.to_csv(path_to_datasets + "/" + 'laplc_train.csv',index=False)
```

TEST Data

```python
len_testset = 1000
```

```python
test_dataset = []
for i in range(len_testset):
    x = np.random.rand()
    y = np.random.rand()
    test_dataset.append((x,y, target_func(x,y)))
```

```python
df_test = pd.DataFrame(test_dataset, columns=['x','y','p(x,y)'])
df_test.to_csv(path_to_datasets + "/" + 'laplc_test.csv', index=False)
```
